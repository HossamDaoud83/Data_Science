```python
# import requests
# from bs4 import BeautifulSoup
# import pandas as pd
# 
# url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
# page = requests.get(url)
# page.raise_for_status()
# soup = BeautifulSoup(page.content, 'html.parser')
# 
# # Try finding the table with different attributes or methods
# # Example: If you're not sure about the ID, you might want to look for any table and check its contents
# tables = soup.find_all('table')
# 
# # If there is more than one table, you might print out their first few rows to identify the right one
# for i, table in enumerate(tables):
#     print(f"Table {i}")
#     rows = table.find_all('tr')
#     for row in rows[:3]:  # Show only first 3 rows for brevity
#         print([td.get_text(strip=True) for td in row.find_all('td')])
# 
# # Assuming you identify the correct table index as 'n'
# correct_table = tables[n]
# 
# # Now, extract rows from the identified correct table
# data = []
# rows = correct_table.find_all('tr')
# for row in rows:
#     cols = row.find_all('td')
#     if cols:
#         data.append([col.text.strip() for col in cols])
# 
# tesla_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])
# print(tesla_revenue.tail())

import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL provided in the request
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
page = requests.get(url)
page.raise_for_status()
soup = BeautifulSoup(page.content, 'html.parser')

# Find all tables, and select Table 0 which contains the annual revenue data
tables = soup.find_all('table')
annual_revenue_table = tables[0]  # Correct table based on the information provided

# Extract rows from the table
rows = annual_revenue_table.find_all('tr')

# Prepare data list for DataFrame
data = []
for row in rows:
    cols = row.find_all('td')
    # Each row should have two columns: Year and Revenue
    if len(cols) == 2:
        year = cols[0].text.strip()
        revenue = cols[1].text.strip()
        data.append([year, revenue])

# Create DataFrame
tesla_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

# Display the DataFrame
print(tesla_revenue)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the webpage
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing GME revenue data
table = soup.find('table')

# Extract data from the table
rows = table.find_all('tr')
data = []
for row in rows:
    cols = row.find_all('td')
    if cols:
        data.append([ele.text.strip() for ele in cols])

# Create a DataFrame
gme_revenue = pd.DataFrame(data, columns=["Year", "Revenue"])

# Display the last five rows of the DataFrame
print(gme_revenue.tail())
```

```python
!pip install yfinance matplotlib
import yfinance as yf
import matplotlib.pyplot as plt
def make_graph(stock_data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(stock_data['Close'])  # Plotting the closing price
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.show()

# Fetch Tesla stock data
tesla_data = yf.Ticker("TSLA").history(period="1y")

# Graph Tesla stock data
make_graph(tesla_data, "Tesla Stock Price - Last Year")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(stock_data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(stock_data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.legend()
    plt.show()

# Fetch GameStop stock data
gme_data = yf.Ticker("GME").history(period="1y")

# Graph GameStop stock data
make_graph(gme_data, "GameStop Stock Price - Last Year")
```
